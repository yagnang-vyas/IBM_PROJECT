## Tasks

Now write and execute SQL queries to solve the assignment tasks.

**Note: If the column names are in mixed case enclose it in double quotes
   For Example "Landing_Outcome"**

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
pd.read_sql('SELECT launch_site, COUNT(launch_site) FROM spacextable GROUP BY launch_site', con)
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
pd.read_sql('SELECT * FROM spacextable WHERE Launch_Site LIKE "CCA%" limit 5', con)
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
pd.read_sql('''SELECT SUM(PAYLOAD_MASS__KG_) AS "Total Payload (kg)" 
            FROM spacextable WHERE Customer LIKE "NASA (CRS)"''', con)
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
pd.read_sql('''SELECT AVG(PAYLOAD_MASS__KG_) AS "Average Payload (kg)"
                  FROM spacextable 
                  WHERE Booster_Version LIKE "F9 v1.1%"''', con)
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
pd.read_sql('''SELECT Date AS "First Date" FROM spacextable 
                  WHERE Landing_Outcome LIKE "Success (ground pad)" 
                  ORDER BY Date LIMIT 1''', con)
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
pd.read_sql('''SELECT Booster_Version FROM spacextable 
            WHERE Landing_Outcome LIKE "Success (drone ship)" 
            AND PAYLOAD_MASS__KG_ BETWEEN 4000 and 6000''', con)
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
pd.read_sql_query('SELECT * FROM spacextable LIMIT 5', con=con)
```

```python
pd.read_sql('''SELECT Mission_Outcome, COUNT(Mission_Outcome) AS "Total" 
            FROM spacextable GROUP BY Mission_Outcome''', con)
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
pd.read_sql('''SELECT Booster_Version FROM spacextable 
            WHERE PAYLOAD_MASS__KG_ = 
                (SELECT MAX(PAYLOAD_MASS__KG_) FROM spacextable)''', con)
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
pd.read_sql('''SELECT Date, substr(Date, 6, 2), Landing_Outcome, Booster_Version, Launch_Site FROM spacextable 
            WHERE Landing_Outcome = "Failure (drone ship)" 
            AND substr(Date, 0, 5) = "2015"''', con)
```

```python
pd.read_sql('''SELECT Date, Booster_Version, Launch_Site FROM spacextable 
            WHERE Landing_Outcome = "Failure (drone ship)" 
            AND substr(Date, 0, 5) = "2015"''', con)
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
pd.read_sql('''SELECT Date, Landing_Outcome, COUNT(Landing_Outcome) AS "Count" 
            FROM spacextable 
            WHERE Date BETWEEN "2010-06-04" and "2017-03-20" 
            GROUP BY Landing_Outcome 
            ORDER BY Count DESC''', con)
```
